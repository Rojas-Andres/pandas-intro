```python
def multiplicacion(id):
    new_id = id * 10
    return new_id 

df["multiplicacion"] = df["id"].apply(multiplicacion)
```

```python
def concatenar(first_name):
    new_first_name = f"{first_name} new"
    return new_first_name

df["concatenar"] = df["first_name"].apply(concatenar)
```
